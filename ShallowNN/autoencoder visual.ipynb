```python
import numpy as np
from autoencoder import Autoencoder
from sklearn import datasets
import matplotlib.pyplot as plt

P=Autoencoder(4,2)
```

```python
C=4 # Classes
X, Y = datasets.make_classification(
    n_features=C,
    n_classes=C,
    n_samples=100,
    n_redundant=0,
    n_clusters_per_class=1
)

Y_=[]
for i in range(len(X)):
    vec=np.zeros((C))
    vec[Y[i]]=1
    Y_+=[vec]
```

```python
P.Train(X,np.array(Y_),200) # Train
```

```python
plt.scatter(X[:,0],X[:,1],c=Y ) # Print points (only 2 coordenates) there are 
```

```python
X_r=P.Encode(X)
plt.scatter(X_r[0],X_r[1],c=Y)
```
